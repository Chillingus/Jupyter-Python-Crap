## Introduction: Coulomb's Law

If two charges $q_1$ and $q_2$ are placed a distance $r$ apart in a vacuum, **Coulomb's Law** states that the force $F$ between them will be directed along the line connecting them, and will have magnitude (in Newtons)  given by:

$$ F = \frac{1}{4\pi\epsilon_0} \frac{q_1 q_2}{r^2}$$

where $\epsilon_0$ is a fundamental physical constant called the *permittivity of the vacuum* and has value $\epsilon_0 = 8.99 \times 10^9$ Nm$^2$/C$^2$.

Whether the force is attractive or repulsive depends on the relative charges: opposite charges will attract, while similar charges will repel.

## Apparatus

This experiment will use a **Coulomb Balance** made by PASCO (Model ES-9070).
The Coulomb Balance involves two identical conducting spheres mounted on rods, one connected to a torsion wire and the other mounted on a slide assembly. 
For reference, the manual for the Coulomb Balance can be found at : https://cdn.pasco.com/product_document/Coulombs-Law-Apparatus-Manual-ES-9070.pdf




The slide assembly has a distance scale marked on it in centimeters, and the torsion wire setup has matching marks on the rotating part and on the base that can be used for alignment.
Before charging the spheres and beginning the experiment, you should verify that the apparatus is aligned correctly: if the distance separation on the sliding assembly is set to 3.8cm, the spheres should be just touching, and the mark on the torsion wire rotating part should line up with the mark on the base. 

The spheres can be charged by applying a voltage from a power source. 
When both spheres are charged, there is an electrostatic force between them. 
The force on the hanging sphere will cause rotation, twisting the torsion wire. The degree of twist is proportional to the force on the sphere, and therefore can be used to measure the force, once the torsion wire is calibrated 
(similar to the Hooke's law relation that the extension of a spring is proportional to the applied force).

The distance between the two spheres can be varied by moving one sphere along its slide assembly, and the amount of charge on the spheres can be varied by changing the applied voltage.

### Tips for Charging the Spheres

A power supply is used to charge the spheres. 
The power supply voltage can be set to a desired value, eg 5kV. When the power supply is on, the sphere can be charged by touching it with a probe connected to the power supply.  

The PASCO User Manual recommends that the spheres be separated as far as possible when charging them, to minimize unwanted additional electrostatic effects. It is also recommended that the power supply be turned off immediately when not being used, again to avoid unwanted electrostatic effects. 
Being mindful also that charge can leak off the spheres, we will recharge them from the voltage supply before each individual measurement, using the following procedure:

1. Turn on the power supply and set it to the desired voltage
2. Turn off the power supply again
3. Slide the sliding part far back to separate the spheres
4. *Discharge* each sphere first by touching with the *ground probe*
5. Turn on the voltage source, and wait a moment to verify the voltage has stabilized
6. Using the charged probe, touch it quickly to each sphere to charge them
7. Turn off the voltage source again quickly
8. Moving the sliding part back to the desired position, and make the measurement as quickly as reasonably possible


### Using the Torsion Wire

Twisting of the torsion wire will be proportional to the force exerted on the attached sphere (like Hooke's law for a stretched spring). 
When there is a force on the sphere, you will see it rotate. Then you will adjust the dial on the torsion wire until it is returned to its initial position (based on lining up the matching marks).
The dial will then show the angle through which it was turned, which will be proportional to the applied force. 




## Experimental Method

For these measurements, we suggest setting the voltage source to 5 kV.
After charging, measure the twist in the torsion wire when the spheres are 20 cm apart. 
Record the distance and twist angle, then

In between each measurement, use the procedure described above to recharge the spheres. By using the same applied voltage each time we recharge, ideally the charge on the spheres will be the same each measurement, so only the distance should be different.

We recommend to perform two separate runs for these measurements, then compare the results to check consistency. 



### Measuring How the Electrostatic Force Between Two Charged Spheres varies with Charge Applied

By changing the magnitude of the voltage when we charge each sphere, we can effectively change the charge applied.
We can assume that the charge applied is proportional to the voltage setting
(the sphere here actually functions like a capacitor, and you will learn later that the charge on a capacitor is proportional to the voltage applied to it).
For this test, we must keep the distance the same each time. We want to use a distance separation where the force will be measurable even at the lowest applied voltage, so probably 6 or 5cm is suggested. 

Use the procedure described above to charge each sphere, then return them to the defined distance separation and measure the torsion wire twist angle for each applied voltage. Try to take measurements at 1,2,3,4 and 5 kV applied voltage. As before, we suggest to do two complete runs for these measurements, then compare the results to check consistency.

```python
import math

dist= [20,15,10,9,8,7,6,5]
angle=[7,13,29,35,41,47,55,67]
angle2=[6,15,28,55,70,85,93,125]

import matplotlib.pyplot as plt

plt.plot(dist,angle,"p")
plt.plot(dist,angle2,"p")


# We expect this relationship to be an inverse-square law (1/dist^2),
# so lets try to fit that to our data
# (Consult the Reference notebook for examples of curve fitting)

#for curve fit
from scipy.optimize import curve_fit

# define the function that we will use to fit
# First argument of this should be the input variable (here generically x, which will the distance from our measurements)
# subseuent arguments are the parameters whose values will be determined by the fit
# For inverse-square law, only free parameter is an overall scale factor A
def inverse_square(x,A):
    return A/(x*x)

# now call curve_fit() with our defined funciton, and the two datasets (corresponding to x and y axes)
# this will return two things: a list with the values of fit parameters ('popt' here)
# and the covariance matrix for the fit parameters ('pcov' below)
# (you dont need to worry too much about the covariance matrix for now ...)

popt,pcov = curve_fit(inverse_square,dist,angle2)

print(popt)
print(math.sqrt(pcov[0]))
x_display = list(range(5,21))
y_display = []
for x in x_display:
    y_display.append(inverse_square(x,*popt))
plt.plot(x_display,y_display,"red")


# lets show that some other functional forms, eg 1/x or 1/x^3 do not fit nearly as well:

def inverse(x,A):
    return A/x

def inverse_cube(x,A):
    return A/(x*x*x)

popt2,pcov2 = curve_fit(inverse,dist,angle)
print(popt2)
print(math.sqrt(pcov2))

popt3,pcov3 = curve_fit(inverse_cube,dist,angle)
print(popt3)
print(math.sqrt(pcov3))


y_display_inverse = []
y_display_inverse_cube = []
for x in x_display:
    y_display_inverse.append(inverse(x,*popt2))
    y_display_inverse_cube.append(inverse_cube(x,*popt3))
    
        
plt.plot(x_display,y_display_inverse,"green")
plt.plot(x_display,y_display_inverse_cube,"purple")

print(angle)

print(angle2)
```
