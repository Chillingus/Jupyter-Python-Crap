## Wavelength Dependence of Refractive Index for Air

The refractive index of air varies slightly depending on the wavelength of light. 
Reference 1 (Ciddor) provides a formula for this dependence.
For air, the dependence is small: when going from blue light at 400 nm wavelength to red light at 650nm, the refractive index decreases by a fraction smaller than $1\times10^{-5}$.


### Example Code: Wavelength Dependence of Refractive Index for Air 

The code below provides a function to evaluate the refractive index of air at a given wavelength (in units of nm). It also plots the refractive index value as a function of wavelength in the range 400-650 nm. 


```python
# example code - defining a function for the refractive index of air at a given wavelength

import math

def refr_index_air(wavel):
    # assume wavelength input in nm -> will need to convert units for the paper's quoted formula
    # formula given in terms of inverse of wavelength, units of inverse micrometers
    sig = 1.0/(wavel*1e-3)
    k0 = 238.0185
    k1 = 5792105
    k2 = 57.362
    k3 = 167917 
    # eqn given is for 1e8*(n_air -1) in Reference paper by Ciddor
    n_air_m1 = 1*(k1/(k0-sig**2)+k3/(k2-sig**2))
    n_air = (1e-8*n_air_m1)+1
    return n_air

# plot the function from 400 to 650 nm
l_range= list(range(400,650,10)) # this range() function creates a set of numbers from 400 to 650, in steps of 10
#print(l_range)

# now eval the refractive index at each of these wavelengths
air_refr_indx_vals = [refr_index_air(l) for l in l_range]
#print(air_refr_indx_vals)

# now plot it. 
# First we import the plotting package
import matplotlib.pyplot as plt
# then we can plot, with the arguments: x,y 
plt.plot(l_range,air_refr_indx_vals)
```

## Deflection of Light Ray Entering Spherical Raindrop

When a light ray enters a raindrop (assumed spherical here), it experiences refraction at the initial surface. Then it can undergo internal reflection at the back end of the raindrop, and finally a second refraction as it exits again. 
The result is a deflection of the light ray relative to its original direction. 
This is illustrated by the diagram below. 




At each of the three surface interactions, there is a deflection $\delta_i$, resulting in a total deflection $\delta = \delta_1 + \delta_2 + \delta_3$ away from the original direction.

By considering the geometry and the laws of reflection/refraction for the interactions, it can be shown that the total deflection angle $\delta$ can be related to the incidence angle $\theta_i$ (shown as $\theta_{1i}$ in the diagram) and the angle of refraction $\theta_r$ at the first air/water interface (shown as $\theta_{1t}$ in the diagram) by:

$$ \delta = 2 (\theta_i - \theta_r) + (\pi - 2\theta_r) $$ 

where all the above angles are in radians.

## A Secondary Rainbow

A secondary rainbow occurs when light rays undergo *two internal reflections* before exiting the raindrop. 
In this situation, the light ray enters into the lower half of the raindrop, undergoes two internal reflections and then emerges again directed downward. 
(Compare to the above diagram for a primary rainbow, where the light ray enters in the upper half of the raindrop, then gets deflected so that is redirected downward, toward an observer on the ground)

In the case of two internal reflections, it can be shown that the deflection angle is given by:

$$ \delta = 2 (\theta_i - \theta_r) + 2(\pi - 2\theta_r) $$ 


(Note that the additional internal reflection also results in a reduction of the light intensity, which is why it's harder to see secondary rainbows in the sky)
