```python
import math
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

def inverse_square(x,A):
        return A/x**2

def inverse(x, A):
       return A/x

def inverse_cube(x, A):
       return A/x**3


dist= [20,15,10,9,8,7,6,5]
angle=[7,13,29,35,41,47,55,67]
angle2=[6,15,28,55,70,85,93,125]

popt,pcov = curve_fit(inverse_square, dist, angle)

x_display = list(range(5, 20))
y_display = []
for x in x_display:
      y_display.append(inverse_square(x, popt))


popt,pcov = curve_fit(inverse, dist, angle)

x_display2 = list(range(5, 20))
y_display2 = []
for x in x_display2:
      y_display2.append(inverse(x, popt))


popt,pcov = curve_fit(inverse_cube, dist, angle)

x_display3 = list(range(5, 20))
y_display3 = []
for x in x_display3:
      y_display3.append(inverse_cube(x, popt))




plt.plot(dist, angle, "o")
plt.plot(dist, angle2, "o")
plt.plot(x_display, y_display, "-")
plt.plot(x_display2, y_display2, "-")
plt.plot(x_display3, y_display3, "-")
plt.show()
print(dist)
print(angle)
print(angle2)


def square(x, A):
    return A * x**2

dist2 = [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4]
angle_2 = [0.5, 3, 6, 16, 19, 34, 31.5, 56]

popt, pcov = curve_fit(square, dist2, angle_2)

x2_display = np.linspace(0, 5, 100)
y2_display = square(x2_display, *popt)

plt.plot(dist2, angle_2, "o")
plt.plot(x2_display, y2_display, "-")
plt.show()
print(dist2)
print(angle_2)
```
