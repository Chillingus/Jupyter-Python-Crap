#### Exercise 1: Wavelength Dependence of Refractive Index for Air

In the cell below, use the function defined above to evaluate the refractive index of air ($n_{air}$) at wavelengths of 400 nm and 650 nm. 
Print your results. Compare the two values, and express the difference as a fraction relative to the value at 400 nm. Confirm the statement above about the size of the change in $n_{air}$ across this range of wavelength.

```python
import math
import matplotlib.pyplot as plt


def refr_index_air(joemama):
    
    sig = 1.0/(joemama*1e-3)
    k0 = 238.0185
    k1 = 5792105
    k2 = 57.362
    k3 = 167917

    n_air_m1 = 1*(k1/(k0-sig**2)+k3/(k2-sig**2))
    n_air = (1e-8*n_air_m1)+1
    return n_air


l_range= list(range(350, 700,10))

air_refr_indx_vals = [refr_index_air(l) for l in l_range]

plt.plot(l_range,air_refr_indx_vals)
plt.title("Refractive index for Air per Wavelength")


gay = 400
result = refr_index_air(400)

gay2 = 650
result2 = refr_index_air(650)

result3 = result2/result

change = result-result2

plt.plot(gay, result, "bo")
plt.plot(gay2, result2, "ro")
plt.xlabel("Wavelength")
plt.ylabel("Refractive index")
plt.show()

print((gay, result))
print((gay2, result2))
print("Red/Blue = " , result3)
print("Change in Refractive Index = ", change)
```

#### The change is indeed smaller than $1\times10^{-5}$

```python
import math
import matplotlib.pyplot as plt

def refr_index_water(x):
    
    n = ((1.31279) + (15.762)/x - (4382)/x**2 + (1.1455*10**6)/x**3)
    return n
 

def refr_index_air(x):
    sig = 1.0/(x*1e-3)
    k0 = 238.0185
    k1 = 5792105
    k2 = 57.362
    k3 = 167917

    n_air_m1 = 1*(k1/(k0-sig**2)+k3/(k2-sig**2))
    n_air = (1e-8*n_air_m1)+1
    return n_air

def calc_deflection(theta_i, x):
    
    
    n_air = refr_index_air(x)
    n_water = refr_index_water(x)
    
    sin_theta_r = math.sin(theta_i) * n_air / n_water
    theta_r = math.asin(sin_theta_r)
    
    D = 2 * (theta_i - theta_r) + (math.pi - 2 * theta_r)
    
    return D*(180/math.pi)


theta_i_values = [math.radians(angle) for angle in range(91)]


Blue = 400
Red = 650

boink_blue = [calc_deflection(theta_i, Blue) for theta_i in theta_i_values]
boink_red = [calc_deflection(theta_i, Red) for theta_i in theta_i_values]

theta_i_degrees = [math.degrees(theta_i) for theta_i in theta_i_values]

Min_Boink_B = min(boink_blue)
Min_Angle_B = boink_blue.index(Min_Boink_B)

Min_Boink_R = min(boink_red)
Min_Angle_R = boink_red.index(Min_Boink_R)


plt.plot(theta_i_degrees, boink_blue, 'b-')
plt.plot(theta_i_degrees, boink_red, 'r-')
plt.title("Deflection of Light Ray in a Raindrop")
plt.xlabel("Incident Angle")
plt.ylabel("Total Deflection Angle")
plt.show()

print("Blue Min Values", (Min_Boink_B, Min_Angle_B) )
print("Red Min Values", (Min_Boink_R, Min_Angle_R) )

observation_angle1 = 90 - Min_Boink_B
observation_angle2 = 90 - Min_Boink_R

print("Angle of Observation for Blue Band", observation_angle1)
print("Angle of Observation for Red Band", observation_angle2)
```

#### The values of the incident angle $\theta_{i}$ that correspond to each minimum deflection angle are $\theta_{i} = 59^{\circ}$ for the Blue wavelength, and $\theta_{i} = 60^{\circ}$ for the Red wavelength. The values of the deflection angle $\delta$ that correspond to each minimum incident angle are $\delta = 139.26^{\circ}$ for the Blue wavelength, and $\delta = 137.56^{\circ}$ for the Red wavelength.


##### In order to see the Blue band you need to look about $49.26^{\circ}$ above the horizon, and in order to see the Red band, $47.56^{\circ}$ above the horizon. Therefore, the Blue band will appear above the Red band. 

#### However, since the observation angle for Blue is greater than that of Red, you will see the Red band above the Blue. Assume there are two water droplets, $D_1$ and $D_2$ both deflect Blue and Red wavelength light rays, but one is above another. $D_1$ is above $D_2$. $D_1$ deflects the Blue ray above the Red ray, and continues onwards missing your eyes, but the Red ray travels down towards your eyes. $D_2$ also deflects the Blue ray above the Red ray at the same deflection angles, but the droplet is lower than $D_1$, thus the Blue rey deflects at a higher angle and reaches your eyes, and the Red ray deflects at a lower angle and travels away from your eyes.
```python
import math
import matplotlib.pyplot as plt

def refr_index_water(x):
    
    n = ((1.31279) + (15.762)/x - (4382)/x**2 + (1.1455*10**6)/x**3)
    return n
 

def refr_index_air(x):
    sig = 1.0/(x*1e-3)
    k0 = 238.0185
    k1 = 5792105
    k2 = 57.362
    k3 = 167917

    n_air_m1 = 1*(k1/(k0-sig**2)+k3/(k2-sig**2))
    n_air = (1e-8*n_air_m1)+1
    return n_air

def calc_deflection(theta_i, x):
    
    
    n_air = refr_index_air(x)
    n_water = refr_index_water(x)
    
    sin_theta_r = math.sin(theta_i) * n_air / n_water
    theta_r = math.asin(sin_theta_r)
    
    D = 2 * (theta_i - theta_r) + 2 * (math.pi - 2 * theta_r)
    
    return D*(180/math.pi)


theta_i_values = [math.radians(angle) for angle in range(91)]


Blue = 400
Red = 650

Boink_blue = [calc_deflection(theta_i, Blue) for theta_i in theta_i_values]
Boink_red = [calc_deflection(theta_i, Red) for theta_i in theta_i_values]

theta_i_degrees = [math.degrees(theta_i) for theta_i in theta_i_values]

Min_Boink_Blue = min(Boink_blue)
Min_Angle_Blue = Boink_blue.index(Min_Boink_Blue)

Min_Boink_Red = min(Boink_red)
Min_Angle_Red = Boink_red.index(Min_Boink_Red)

plt.plot(theta_i_degrees, Boink_blue, 'b-')
plt.plot(theta_i_degrees, Boink_red, 'r-')
plt.title("Deflection of Light Ray in a Secondary Raindrop")
plt.xlabel("Incident Angle")
plt.ylabel("Total Deflection Angle")
plt.show()


print("Blue Min Values :", (Min_Boink_Blue, Min_Angle_Blue))
print("Red Min Values :", (Min_Boink_Red, Min_Angle_Red))

obsB = 90 - Min_Boink_Blue
obsR = 90 - Min_Boink_Red

print("Angle of Observation for Blue Ray :", obsB)
print("Angle of Observation for Red Ray :", obsR)
```

## Extra Credit: Maximum Light Intensity at Minimum Deflection Angle

To predict the rainbow angles in the sky from our model, we assumed that the peak intensity of observed light rays corresponded to the position of *minimum deflection angle*.
And indeed this is the case, but why? 
Light enters the raindrops across the whole range of incidence angles. 
Assume when we look up at rainbow, that we are seeing light from a small window $\Delta \theta$ in observed angle. 
**For extra credit**, can you give a brief rationale why you might observe more light intensity for a particular color close to the minimum deflection angle for that wavelength than at any random angle?

#### Light rays entering the raindrop are focused towards the back of the raindrop due to refraction, the focusing is more pronounced for light rays that are closer to the minimum deflection angle because they undergo less deviation from their original path. At the back of the raindrop, the focused light rays interfere with each other constructively and destructively. Constructive interference occurs when the path length of the light rays is nearly the same, reinforcing the intensity of light. Destructive interference occurs when the path length of the light rays differs significantly, reducing the intensity of light. Light rays that are closer to the minimum deflection angle have a higher probability of undergoing constructive interference at the back of the raindrop, leading to a peak in intensity for that specific color band at that angle.
