## Introduction

Here we briefly summarize the theory behind reflection and refraction in optics. 
Further explanation can be found in the introductory physics textbook. 

### Law of Reflection

When light is reflected off a surface, the reflected light ray is related to the incident light ray through the Law of Reflection which states that: *the angle of incidence is equal to the angle of reflection*.
Both angles are measured relative to the normal of the surface causing the reflection.

### Snell's Law of Refraction

When an incident light ray enters a different medium, part of the light is reflected at the surface (this part obeys the law of reflection, as above), and part travels through the new medium, but with a different angle compared to the incident ray. For the light that travels through the new medium, we call this the *refracted ray*.
Measured relative to the normal of the surface separating the two media, the angle of the light beam in the first medium $\theta_1$ is related to the angle of the refracted beam $\theta_2$ by Snell's law of refraction: 
$$ n_1 \sin\theta_1 = n_2 \sin\theta_2 $$
where $n_1$ and $n_2$ are dimensionless constants called the *refractive index* of each medium. It can be shown that the refractive index $n$ of a material is related to the speed of light in the material ($v$) relative to the speed of light in a vacuum $c$, by $n=v/c$

### Total Internal Reflection

When a light ray passes from a more optically-dense (larger refractive index $n$) into a less optically-dense (smaller $n$) medium (an example would be a light ray coming out of water and into air), then the light is refracted according to Snell's law above. 
However, if the incidence angle between the light ray and the normal to the medium surface is large enough, then one can observe that there is a certain point when the angle of the refracted ray can reach 90 degrees - in this situation, the refracted ray actually does not leave the first medium, it instead propagates along the surface separating the media. We call this the *critical angle* $\theta_c$, and for any light rays which have incident angles $\theta > \theta_c$, there is no refracted ray at all, and all the light is reflected back into the original medium. We call this phenomenon *total internal reflection*.

We can obtain an expression for this critical angle by substituting a specific angle of 90$^\circ$ for the refracted ray into Snell's law of refraction. Thus we have $$ n_1 \sin \theta_c = n_2 \sin 90^\circ$$
and then $$\sin \theta_c = n_2/n_1$$
Note that we must have $n_1>n_2$ in order for total internal reflection to happen.

## Extra Credit: Wavelength Dependence of Refractive Index

On a more detailed level, the refractive index of a material can actually vary depending on wavelength, because light rays with different wavelengths can propagate at different speeds within the medium. Typically, for a shorter wavelength (say, blue light) the refractive index of the material will be greater than for a longer wavelength (say red light). This is often called *dispersion*, because light of different wavelengths will spread out with different refraction angles in a medium. This effect is what causes the splitting of light into a rainbow, as different visible colors correspond to different wavelengths of light. 
(We will explore the phenomenon of a rainbow in more detail in a future lab.)
