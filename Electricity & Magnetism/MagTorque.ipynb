```python
import math
import numpy as np
from matplotlib import pyplot as plt


I = [2.1, 2.4, 2.6, 2.7, 2.9, 3.2]
d = [0.04345, 0.05345, 0.06345, 0.07345, 0.08345, 0.09345]
b = 0.001356
T_w = [0.0577, 0.0710, 0.0843, 0.0976, 0.1109, 0.1242]  


for i in range(len(I)) and range(len(d)) and range(len(T_w)):

        B = (b)*(I[i])


        

        mu = (T_w[i])/(B)

        print("Magnetic Dipole Moment:", mu)
        print("Magnetic Field:", B)

        
plt.plot(d, I, 'o')


fit_coeffs = np.polyfit(d, I, 1)
fit_line = np.poly1d(fit_coeffs)

slope = fit_coeffs[0]
intercept = fit_coeffs[1]

print(f"Equation of the line of best fit: y = {slope:.4f}x + {intercept:.4f}")


plt.plot(d, I, 'ro', label='Data')

plt.plot(d, fit_line(d), color='blue', label='Line of Best Fit')
```
