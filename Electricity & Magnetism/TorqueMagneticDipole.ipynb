# Torque on a Magnetic Dipole


## Overview

A dipole magnet (such as a bar magnet or a current loop) has a magnetic dipole moment that causes it to experience a torque when placed in an external magnetic field. In this lab we will measure this torque in order to determine the magnetic dipole moment of an unknown magnet. 


## Introduction

A force $\vec{F}$ can cause rotation of an object when that force is applied at a point that is some distance from an axis of rotation, and the force has some component that it is perpendicular to the line connecting the axis of rotation with the point of force application. 
The *torque*  $\vec{\tau}$ is defined as: $\vec{\tau} = \vec{r} \times \vec{F}$, where $\vec{r}$ is the displacement vector from the axis of rotation to the point where the force $\vec{F}$ is applied.
The direction of rotation produced by the torque is given by the Right-hand rule. 
(Note that the object does not actually rotate in the direction of the torque vector itself; rather it rotates under the influence of the force vector that produces the torque.)

When a *magnetic dipole moment* $\vec{\mu}$ is placed in some external magnetic field $\vec{B}$, it experiences a torque $\vec{\tau}$ given by:  $ \vec{\tau} =  \vec{\mu} \times \vec{B} $


Conceptual question: what way does this magnetic torque try to align the dipole moment relative to the field direction?
Discuss this with each other and the TA/instructor as needed.
Note that vertically down (relative to $\vec{B}$) is also a zero-torque position, but this would  not be *stable* equilibrium, as any additional force would cause it to move upward and away, rather than restoring to the downward position.


Thus if we know the external magnetic field $\vec{B}$ and can measure the magnetic torque (as will be discussed below), then we can determine $\vec{\mu}$ for the unknown magnet dipole moment. 


## Experimental Setup

The apparatus used for this experiment is shown in the image below.




The billiard ball in the center has a permanent dipole magnet embedded inside it. This is the magnet whose magnetic dipole moment $\vec{\mu}$ we are going to measure. 
The central support column allows air to be pumped upwards through it, supporting the billiard ball on an air cushion and allowing the ball to move freely under the influence of the external magnetic field, with minimal friction. 

The external magnetic field is produced by the coil arrangement (similar but not identical to the Helmholtz coil that we have used in previous labs). 
The magnetic field orientation should be directed upwards in this experiment.
The magnetic field strength can be varied by adjusting the current flowing through the coils, which in turn changes the torque experienced by the magnetic dipole inside the billiard ball.  (The formula relating the magnetic field strength of the coils to the current flow is given below.) 


The billiard ball has a 'handle' that allows a rod to be attached. 
The internal magnetic dipole inside the ball is orientated parallel with the handle where the rod is inserted into the ball; thus the upward external magnetic field in this experiment will exert a *magnetic torque* that tries to orient the ball such that the rod points *upward*. 



A weight is placed on the rod. This creates an opposing torque due to the weight; when the rod is horizontal, the gravitational torque from the weight has magnitude:
$$\tau_W = rmg$$
where $m$ is the mass of the weight, $r$ is the distance of the weight location from the center of the ball, and $g=9.81$m/s$^2$ is the acceleration due to gravity.

The gravitational torque will try to rotate the ball such that the rod points downward.
**If the gravitational torque is cancelled out by the magnetic torque, then the rod will remain horizontal** - this is the principle that we will use in this experiment.

The weight can be moved along the rod to different positions, varying $r$ in the above formula. We will then adjust the current $I$ in the coil to change the magnetic field such that the magnetic torque on the dipole magnet cancels out the gravitational torque. This will allow us to calculate the magnetic dipole moment of the embedded magnet. 

(Note that rod itself has some mass, which acts through its center-of-mass, and this will create additional gravitational torque. 
However, since this will not change during this experiment, it will not affect our results provided we account for it correctly. )



### Magnetic Field of Coil Apparatus

The coil apparatus used here is similar but not identical to the Helmholtz coil arrangement used in previous labs. 
There are two coils, each with $N$ turns and radius $R$, separated by a distance $d$.
In this case, the magnetic field at the center of the apparatus ($z=0$) can be shown to be:
$$ B(z=0) = \frac{\mu_0 N I R^2}{\left( R^2 + \frac{d^2}{4} \right)^{3/2}} $$

For our specific apparatus, we have the values: N = 195 turns in each coil; coil radius $R=0.109$ m; separation $d=0.138$ m. 
This results in a current-dependent magnetic field at the center of the apparatus given by:
$$ B(I) = 1.356 \times 10^{-3} I~~\textrm{Tesla}$$ 

where $I$ is the current in Amperes.

## Measurements to Make

First measure the mass $m$ of the movable weight. 

Start with the weight as close as possible to the center of the ball, and measure the distance with Vernier calipers. Turn on the air flow, and start to adjust the current in the coils until the magnetic torque cancels the gravitational torque and the rod is held horizontally. Record this current. 

*(Note that it can be very tricky to get the rod perfectly horizontal. This is because if there is even a small difference between the opposing torques, then the net torque will drive the  rod very quickly toward the upward/downward direction. However, if you can identify the current value at which the rod quickly changes from downward to upward, that will be sufficient for this experiment. You should be able to determine this current value to an accuracy of about 0.1A)*

Now turn the apparatus off, and set the weight to a new distance along the rod, maybe about 1 cm further out. Measure this distance, then repeat the above steps and record the new current that is needed in the coils for the magnetic torque to cancel the gravitational torque. 

Repeat this process for several different distances, making 6-8 measurements of the weight's distance and the corresponding coil currents. Currents in the range of 2-4 A will be needed to create sufficient magnetic torque. Note also that if the weight is placed too far away (eg at the very end of the rod), then there may not be sufficient current available from the power supply to make a large enough opposing magnetic torque.
