## Overview


A **capacitor**  is a device for storing electrical energy/charge. 
In this lab, we will study the discharging of capacitors with a resistor, and by fitting to the discharge data, determine the time constant for the RC circuit. 
We will also study combinations of capacitors arranged in series or in parallel, and attempt to verify that the effective capacitances of these systems agrees with the theoretical predictions. 

## Introduction


Unlike a conductor, electrical current does not flow smoothly through a capacitor, so when a voltage is applied to a capacitor, charges pile up in different regions, creating a charge imbalance. 
One type of capacitor that is easy to visualise is a parallel-plate capacitor, consisting of two metal plates arranged parallel to each other, but separated so that current cannot flow directly across the gap between plates. In this capacitor, because current cannot flow across the gap, applying a voltage will cause a charge $+q$ to build up on one plate, and $-q$ to build up on the other plate. 
If the applied voltage is $V$, then the **capacitance** $C$ of a capacitor is defined such that the magnitude of charge $q$ that builds up on one plate is $q=CV$ 

(Note that the total charge is still zero, because the -q and +q regions cancel each other out, since charge is conserved. So we only consider the magnitude of charge on one plate)


### Discharging a Capacitor

When the applied voltage is turned off, we are left with a charge imbalance across the capacitor, and as a result, current starts to flow back in the opposite direction, trying to equalize again and discharge the capacitor. 

Suppose the capacitor is discharging through a resistor $R$. 
As the capacitor discharges, it generates a time-dependent voltage $V(t)$ based on its capacitance $C$ and the *charge remaining on the plate* at the time $t$, $q(t)$: 

$$ V(t) = q(t)/C$$

At time $t$, the current through the resistor $I(t)$ is proportional to the voltage across the resistor at that time $V(t)$, from Ohm's Law: $I(t) = V(t)/R$.
And the current itself is simply the rate of change of charge: $I(t) = \frac{dq}{dt}$

So we can write an equation for the discharge of capacitor $C$ through resistor $R$:

$$ \frac{dq}{dt} = \frac{V(t)}{R} = -\frac{q(t)}{RC}$$

(Note that we included a minus sign here because the charge is decreasing as the capacitor discharges)
<br>
We can integrate this equation to find that the charge remaining on the capacitor at time $t$, $q(t) = q_0 \exp(-t/RC)$
where $q_0$ was the initial charge of the fully-charged capacitor. 
It's not so easy to measure charge directly, but we can obtain from this an expression for the voltage across the capacitor as a function of time, $V(t)$, and this is something we can measure with a voltmeter. 
Using $q(t) = C V(t)$, we obtain that the voltage across a capacitor $C$ discharging through a resistor $R$ is given by:
$$ V(t) = V_0 \exp(-t/RC) $$
The quantity $RC$ is called the **time constant** of the RC circuit; it measures how quickly the capacitor charges or discharges.


### Capacitors Connected in Parallel in a Circuit

Now let's consider two capacitors $C_1$ and $C_2$ connected in parallel, and this capacitor system is then connected in series to a resistor $R$.
These two capacitors will behave effectively like a single capacitor $C_p$, but what is the capacitance of $C_p$?

Since the capacitors are in parallel, the same voltage $V$ will get applied to both. But since their capacitances are different, that means the charge built up on each will be different: $q_1 = C_1 V$ and $q_2= C_2 V$.

Thus the total charge built up on the system will be $q_{tot} = q_1 + q_2$.
Therefore the effective capacitance $C_p$ of the system will be related to the total charge $q_{tot}$ and the applied voltage $V$ by: 
$$q_{tot} = C_p V = q1 + q2 = C_1V + C_2V = (C_1+C_2) V$$
From this we can determine that the effective capacitance of capacitors connected in parallel is:
$$ C_p = C_1 + C_2$$


### Capacitors Connected In Series in a Circuit

Finally, let's consider two capacitors $C_1$ and $C_2$ connected in series, along with a resistor $R$.
The two capacitors will behave effectively like a single capacitor $C_s$, but what is the capacitance of $C_s$?

In this case, there will be a different voltage $V_1$ and $V_2$ across each capacitor, with the total equalling the applied voltage $V = V_1+V_2$. 

Now we must consider how the charge builds up on the capacitors, to figure out $V_1$ and $V_2$ separately. 
First imagine the charge flowing and reaching the *first plate* of the capacitor $C_1$, and then it will stay there, because it cannot flow further. Let's say an amount $q$ builds up on that plate.
Then, on the opposite end of the circuit, to conserve charge, an amount $-q$ must build up on the *second plate* of the second capacitor $C_2$.
 So now we just need to figure out the charge on the second plate of C1 and the first plate of C2.
 Now, a charge $q$ on the first plate of C1 will repel charge away from it, so the result is that the second plate of C1 will end up with a charge of $-q$. And then to conserve charge, the first plate of C2 (which is connected to the second plate of C1), will end up with a charge $+q$.
 So there will be a charge $q$ on each capacitor, meaning that they will have a voltage:
 $V_1 = q/C_1$ and $V_2 = q/C_2$
 
 So we can see that the combined set of series capacitors will look like a single capacitor, with charge $q$, applied voltage $V$ and effective capacitance $C_s$ related by:
 $$ q = C_s V = C_s (V_1 + V_2) = C_s (\frac{q}{C_1} + \frac{q}{C_2}) = q C_s(\frac{1}{C_1} + \frac{1}{C_2}) $$
 Thus we find that:
 $$ \frac{1}{C_s} = \frac{1}{C_1} + \frac{1}{C_2} $$
