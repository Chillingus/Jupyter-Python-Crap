# Measuring the Earth's Magnetic Field

## Overview

In this lab we will measure the Earth's magnetic field, by studying the deflection of a compass needle when an additional generated magnetic field is applied perpendicular to the Earth's field. 


## Introduction


The Earth's magnetic field is approximately that of a dipole (bar) magnet, with one pole close to the geographic North Pole, and the other close to the geographic South Pole.

We can determine the direction of the Earth's magnetic field using a compass. 
A compass is simply a small dipole magnet that is free to rotate, such that it will align itself with the field lines due to an external magnetic field. The compass needle will point in the direction of *magnetic North*, which is not quite the same as the geographical North Pole
(because the Earth's magnetic field axis is displaced by about 11 degrees relative to the Earth's rotation axis).

The compass tells us the direction of the magnetic field that it is in,  but it 
cannot tell us the *strength* of this field. 
However, if we were to generate an additional known magnetic field near the compass, **the compass would now point in the direction of the total field that results from the vector sum of the Earth's magnetic field plus the additional field**.
In that case, by measuring the angle of deflection of the compass, we could use our knowledge of the additional field to infer the strength of the Earth's magnetic field alone. 

To produce the additional known magnetic field, we will use an apparatus called a 
*Helmholtz coil*.
A Helmholtz coil is a simple arrangement to produce a reasonably-constant magnetic field in the center of the apparatus. 
Two wire coils, each of $N$ turns and radius $R$, are placed parallel to each other and separated by a distance also equal to $R$ along what we will choose to be the $z$-axis. 
A current $I$ flows in both coils, and at the center of the apparatus ($z=0$), the magnetic fields from both coils point in the same direction ($\pm \hat{z}$, depending on the direction of the current flow). 
It can be shown that the magnitude of the magnetic field in the center of the apparatus is given by:
$$ B_z = \frac{8 \mu_0 NI}{5 \sqrt(5) R} $$

where $\mu_0=4\pi \times 10^{-7}$ N/A$^2$ is the permeability of free space. 


## Experimental Setup

The apparatus that we will use is shown in the photograph below, with a compass needle located in the center of a Helmholtz coil.
We will vary the current flowing through a Helmholtz coil and measure the resulting deflection angle of the compass needle.
From this data, we will be able to calculate a value for the strength of the Earth's magnetic field. 





You should ensure that both coils of the Helmholtz coil apparatus are connected to the power supply, and that the current direction of flow in each coil is such that the generated magnetic fields will point in the same direction. A multimeter should be connected to read the applied current. 

With the power supply to the Helmholtz coil turned off, 
position the compass needle in the center of the apparatus. 
Now align the axis of the Helmholtz coil until it is **perpendicular** to the compass needle direction. 
Thus, the roughly-constant field from the Helmholtz coil will be at 90$^o$ to the Earth's magnetic field that points to magnetic North
(this is the 'horizontal' component of the Earth's magnetic field; as we will discuss later, there is also a vertical component to the Earth's magnetic field).


Now turn on the power to the Helmholtz coil. The total magnetic field will then be the vector sum of the contributions from the Helmholtz coil and from the Earth's field, which should be orthogonal to one another. 
As you adjust the applied current, you should see the compass needle move under the influence of the total magnetic field.

## Extra Credit: Measuring the Vertical Component of the Earth's Magnetic Field

Our measurements above will allow us to determine a value for the horizontal component of the Earth's magnetic field.
The Earth's magnetic field also has a *vertical* component, which in the Northern hemisphere is directed down into the Earth  (imagine standing right on top of the magnetic North pole -- then a compass would point directly *down*).
A *Dip Needle* is a compass mounted on a plane that can itself be rotated between horizontal and vertical alignments. 

Once you have determined the horizontal component of the Earth's magnetic field, for extra credit you can consider
how might you use a Dip Needle to determine the vertical component.
