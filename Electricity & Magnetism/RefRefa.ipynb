```python
import math
import numpy as np
from matplotlib import pyplot as plt


A_Refra_1 = [31, 45, 62, 71.5, 51.5]

A_Refra_2 = [55, 61.5, 71, 77, 65.5]

sin_refra_values = []
sin_in_values = []

for i in range(len(A_Refra_1)) and range(len(A_Refra_2)):

        Sin_refra = math.sin(((math.pi)/(180))*(A_Refra_2[i]))

        Sin_in = math.sin(((math.pi)/(180))*(A_Refra_1[i]))

        sin_refra_values.append(Sin_refra)
        sin_in_values.append(Sin_in)

        print("Refract", Sin_refra)
        print("In" , Sin_in)

        plt.plot(Sin_in, Sin_refra, 'bo')


coefficients = np.polyfit(sin_in_values, sin_refra_values, 1)
poly = np.poly1d(coefficients)
plt.plot(sin_in_values, poly(sin_in_values), color='red')


print("Equation of the line of best fit:")
print("y =", coefficients[0], "x +", coefficients[1])

```
