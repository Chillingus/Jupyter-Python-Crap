```python
import numpy as np
import matplotlib.pyplot as plt

# Define a probability density function for the electron position
def electron_pdf(x, y):
    # Example PDF: a Gaussian distribution centered at (0, 0) with std deviation of 1
    return np.exp(-(x**2 + y**2) / 2)

# Generate random samples from the PDF using the rejection sampling method
def generate_samples(pdf, num_samples, bounds):
    samples = []
    while len(samples) < num_samples:
        x = np.random.uniform(bounds[0][0], bounds[0][1])
        y = np.random.uniform(bounds[1][0], bounds[1][1])
        u = np.random.uniform(0, 1)
        if u < pdf(x, y):
            samples.append((x, y))
    return np.array(samples)

# Parameters
num_samples = 10000
bounds = [(-3, 3), (-3, 3)]  # Bounds for x and y

# Generate samples
samples = generate_samples(electron_pdf, num_samples, bounds)

# Scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(samples[:, 0], samples[:, 1], s=1, alpha=0.5)
plt.title('Probability Density Electron Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def electron_pdf(x, y, z):

    return np.exp(-(x**2 + y**2 + z**2) / 2)

def generate_samples(pdf, num_samples, bounds):
    samples = []
    while len(samples) < num_samples:
        x = np.random.uniform(bounds[0][0], bounds[0][1])
        y = np.random.uniform(bounds[1][0], bounds[1][1])
        z = np.random.uniform(bounds[2][0], bounds[2][1])
        u = np.random.uniform(0, 1)
        if u < pdf(x, y, z):
            samples.append((x, y, z))
    return np.array(samples)


num_samples = 10000
bounds = [(-3, 3), (-3, 3), (-3, 3)]


samples = generate_samples(electron_pdf, num_samples, bounds)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(samples[:, 0], samples[:, 1], samples[:, 2], s=1, alpha=0.5)
ax.set_title('Probability Density Electron Scatter Plot (3D)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
