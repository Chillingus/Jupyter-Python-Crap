```python
import numpy as np
import csv
import matplotlib.pyplot as plt

#FIRST DROP

with open('70g.csv') as file:
    reader = csv.reader(file)
    Time = []
    Position = []
    for row in reader:
        Position.append(row[1])       
        Time.append(row[0])

Time = Time[4:]
Position = Position[4:]

Position = [float(i) for i in Position]

# Convert position data to distance fallen


Time_Physical = [float(i) - float(Time[1]) for i in Time]

#p = np.polyfit(Time_Physical, Position, deg=2)
#y = np.polyval(p, Time_Physical)

print()
plt.plot(Time_Physical, Position, "-")
#plt.plot(Time_Physical, y)
plt.xlabel('Time (s)')
plt.ylabel('Distance Fallen (m)')
plt.show()
```
