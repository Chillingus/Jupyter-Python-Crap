## Simple Harmonic Motion

When an object is displaced from equilibrium and experiences a *restoring force*, that results in oscillations known as **simple harmonic motion** (SHM). A classic example of this is the oscillation of a mass on a spring. 
When the spring is displaced a distance $x$ from equilibrium, the mass $m$ experiences a Hooke's Law force $-kx$, where $k$ is the spring constant. 
This is a *restoring force* because (a) it is always directed back toward the equilibrium position; and (b) its magnitude increases linearly in proportion to the displacement. 
The resulting motion of the mass can be described by a sinusoidal function of the time. In the case where the mass begins its motion with at rest (zero initial velocity) at an amplitude $A$ from the equilibrium position, the position at any later time $t$ is: 

$$ x(t) = A \cos \omega t $$

where $\omega$ is the *angular frequency of oscillation*, in units of radians/sec. 
For motion that does not start specifically at an amplitude 
(for example, the mass begins at the equilibrium position but is given some non-zero initial velocity),
the position function can be described with an additional phase angle $\phi$ eg $A\sin(\omega t + \phi)$. 

By considering Newton's Second Law ($F=ma$) applied to the mass $m$ displaced by some distance $x$ from equilibrium, it can be shown that the angular frequency of oscillation depends on the spring constant $k$ and $m$ via:

$$ \omega ^2 = k/m$$

We will test this relation in the experiments below, by observing $\omega$ for different applied masses on an oscillating spring. 
We will first need to measure the spring constant $k$.

Now, if we look again at the theoretical expression for the angular frequency of oscillation, we note that it depends on $k$ and $m$, but does not involve at all the *amplitude* of the motion, $A$. This implies that provided $k$ and $m$ do not change, then oscillations with different amplitude $A$ should still have the same frequency. 
This may at first seem surprising, because with a larger amplitude, the object has a larger distance to travel in each cycle. 
But the key is that the Hooke's law force form the spring also increases in magnitude with larger displacements, so these effects cancel each other out.  This amplitude-independence of the oscillation frequency is thus a key prediction of SHM - so let's test it.

## Introduction to Damping

Simple harmonic motion occurs when the only force present is the restoring force, such as the Hooke's law force from a stretched spring. 
An additional velocity-dependent force can cause an effect of *damping* the motion -- the system still oscillates, but the amplitude of oscillation is observed to decay over time. You will learn more about damped oscillations at a later stage in your physics education, but here we will take a quick experimental look at this concept. 

We can make a first study of damping in our setup by increasing the air drag on the oscillating mass.
We do this by placing an old CD on top of the mass hanger - this increases the surface area of the system, and now when the mass oscillates, air resistance on the larger surface area introduces a damping force.
