```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from tabulate import tabulate

with open('Pressure Lab.csv') as file:
    reader = csv.reader(file)
    Volume = []
    Pressure = []
    for row in reader:
        Pressure.append(row[1])       
        Volume.append(row[0])

Volume = Volume[1:]
Pressure = Pressure[1:]

Pressure = [float(i) for i in Pressure]

plt.plot(Volume, Pressure, "-")

plt.xlabel('Volume (m^3)')
plt.ylabel('Pressure (Pa)')

plt.xticks(rotation=45)

table = list(zip(Volume, Pressure))

headers = ["Volume (m^3)", "Pressure (Pa)"]
table_str = tabulate(table, headers, tablefmt="fancy_grid")

print(table_str)

plt.show()
```

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

with open('Pressure Lab.csv') as file:
    reader = csv.reader(file)
    Volume = []
    Pressure = []
    for row in reader:
        Pressure.append(row[1])       
        Volume.append(row[0])

Volume = Volume[1:]
Pressure = Pressure[1:]

Pressure = [1/float(i) for i in Pressure]

Volume = [float(i) for i in Volume]

plt.plot(Volume, Pressure, "o")

table = list(zip(Volume, Pressure))

plt.xlabel('Volume (m^3)')
plt.ylabel('Pressure (Pa)')

Volume = np.array(Volume, dtype=float)
Pressure = np.array(Pressure, dtype=float)

fit_params = np.polyfit(Volume, Pressure, 1)

fit_line = np.polyval(fit_params, Volume)

plt.plot(Volume, fit_line, "--", label="Line of Best Fit")

plt.legend()
print(table_str)
plt.show()

print("Gradient of the line of best fit:", fit_params[0])
print("Intercept of the line of best fit:", fit_params[1])
```
