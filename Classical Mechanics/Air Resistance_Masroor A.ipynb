# Finding the Terminal Velocity from the Data

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

# FIRST DROP

with open('Weighted Analysis.csv') as file:
    reader = csv.reader(file)
    Time = []
    Position = []
    for row in reader:
        Position.append(row[1])       
        Time.append(row[0])

Time = Time[6:]
Position = Position[6:]

Position = [float(i) for i in Position]

# Convert position data to distance fallen
initial_position = Position[0]
Distance_fallen = [initial_position - float(pos) for pos in Position]

Time_Physical = [float(i) - float(Time[1]) for i in Time]

# Getting terminal velocity
Velocity = np.gradient(Distance_fallen, Time_Physical)

# Find acceleration to find at what point acceleration flattens out
Acceleration = np.gradient(Velocity, Time_Physical)

p = np.polyfit(Time_Physical, Distance_fallen, deg=2)
y = np.polyval(p, Time_Physical)

print(p)

# Plot Acceleration over time to find the terminal velocity interval
plt.plot(Time_Physical, Acceleration, label="Acceleration")
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.title('Acceleration vs Time')
plt.show()

# Terminal velocity interval
Terminal_Velocity_Interval = slice(0, None)
# Avg terminal velocity
Average_Terminal_Velocity = np.mean(Velocity[Terminal_Velocity_Interval])

print(f"Average Terminal Velocity: {Average_Terminal_Velocity} m/s")


plt.plot(Time_Physical, Distance_fallen, "-", label="Distance Fallen")
plt.plot(Time_Physical, y, label="Fitted Curve")
plt.xlabel('Time (s)')
plt.ylabel('Distance Fallen (m)')
plt.title('Distance Fallen vs Time')
plt.show()
```

### Finding Terminal Velocity for all Data sets:

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

def analyze_csv(file_name):
    with open(file_name) as file:
        reader = csv.reader(file)
        Time = []
        Position = []
        for row in reader:
            Position.append(row[1])       
            Time.append(row[0])

    Time = Time[6:]
    Position = Position[6:]

    Position = [float(i) for i in Position]

    # Convert position data to distance fallen
    initial_position = Position[0]
    Distance_fallen = [initial_position - float(pos) for pos in Position]

    Time_Physical = [float(i) - float(Time[1]) for i in Time]

    # Getting terminal velocity
    Velocity = np.gradient(Distance_fallen, Time_Physical)

    # Find acceleration to find at what point acceleration flattens out
    Acceleration = np.gradient(Velocity, Time_Physical)

    # Terminal velocity interval
    Terminal_Velocity_Interval = slice(0, None)
    
    # Avg terminal velocity
    Average_Terminal_Velocity = np.mean(Velocity[Terminal_Velocity_Interval])

    

    return Time_Physical, Distance_fallen, Velocity, Acceleration

# List of CSV files
csv_files = ['Weighted Analysis.csv', 'Unweighted Analysis.csv', 'Plastic Analysis.csv', 'Foam Analysis.csv']

# Plot all distance fallen vs time on one graph
plt.figure()
for csv_file in csv_files:
    time, distance, _, _ = analyze_csv(csv_file)
    plt.plot(time, distance, label=f"{csv_file} - Distance Fallen")

plt.xlabel('Time (s)')
plt.ylabel('Distance Fallen (m)')
plt.title('Distance Fallen/Time')
plt.legend()
plt.show()

# Plot all acceleration on one graph
plt.figure()
for csv_file in csv_files:
    time, _, velocity, acceleration = analyze_csv(csv_file)
    plt.plot(time, acceleration, label=f"{csv_file} - Acceleration")

plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.title('Acceleration/Time')
plt.ylim(bottom=-10)
plt.legend()
plt.show()


# Plot all Velocities on one graph
plt.figure()
for csv_file in csv_files:
    time, _, velocity, _ = analyze_csv(csv_file)
    plt.plot(time, velocity, label=f"{csv_file} - Velocity")

plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity/Time')
plt.legend()
plt.show()

# Average Terminal Velocity for each file
for csv_file in csv_files:
    _, _, velocities, _ = analyze_csv(csv_file)
    # Terminal velocity interval
    Terminal_Velocity_Interval = slice(0, None)
    # Avg terminal velocity
    Average_Terminal_Velocity = np.mean(velocities[Terminal_Velocity_Interval])

    print(f"Average Terminal Velocity for {csv_file}: {Average_Terminal_Velocity} m/s")
```

### Finding the Drag Coefficient

$v_T = \sqrt{\frac{mg}{c \rho A}}$

```python
# Calculate c for each set of values
import math
import numpy as np

def solve_for_c(m, g, rho, A, v_T):
    c = (m * g) / (rho * A * v_T**2)
    return c

# Example usage with arrays for m, A, and v_T
mass_array = np.array([274.5, 51.9, 19.1, 11.2])  # array of masses in kg
cross_sectional_area_array = np.array([39, 39.5, 31.17, 64.326])  # array of cross-sectional areas in m^2
terminal_velocity_array = np.array([5.455129813119278, 4.933560591041689, 4.866971126985394, 4.209611479146158])  # array of terminal velocities in m/s

gravity = 9.8  # acceleration due to gravity in m/s^2
air_density = 1.2  # air density in kg/m^3

result_c_array = solve_for_c(mass_array, gravity, air_density, cross_sectional_area_array, terminal_velocity_array)
print("Drag Coefficient (c) for each set of values:", result_c_array)
```
