## Objective

In this lab we will test the hypothesis that angular momentum is conserved in changes of motion, provided that there are no external torques acting. 
We will use a spinning system where the distribution of mass (and hence the rotational inertia) will change mid-motion. By determing the rotational inertia values and measuring the angular speeds before and after the change, we will test if the angular momentum of the system is conserved through the change.

The setup will use a rotating chair, with an attached barbell holding movable weights. A person sitting in the chair can move the weights to an inner or an outer position; the forces from this movement should be directed toward/away from the axis of rotation, and therefore they should generate no torque. In this way we will change the rotational inertia of the spinning system without any external torque, and then test if the angular momentum is conserved. We will also test if kinetic energy is conserved as well.

So the steps in brief are:

1. Determine the before/after rotational inertia values for the system, with the weights in the inner/outer positions. This is done by studying small oscillations of the system.
2. Start the system rotating with the weights in an initial position, and measure the initial angular velocity
3. Change the position of the weights along the barbell *while in motion*
4. Record the final angular velocity
5. From the initial and final angular velocities, and the before/after rotational inertia values, determine if angular momentum is conserved.
6. Finally consider from the data if *rotational kinetic energy* is conserved as well


## Rotating Objects, Rotational Inertia and Angular Momentum

For rotating objects, the *rotational inertia* $I$ (defined below), plays an equivalent role as mass $m$  does for non-rotating motion. Similarly, the *angular velocity* $\omega$, defined as the rate of change of the angular position of an object, plays the role of linear velocity $v$.
It can be shown that the *angular momentum* $L$ of a rotating object is related to $I$ and $\omega$ by: $$ L = I\omega$$
(Note the similarity to the linear momentum expression $p=mv$, with the idea that $I$ 'replaces' $m$ and $\omega$ 'replaces' $v$)

The rotational inertia $I$ of an object depends on its mass, but also on how this *mass is distributed relative to the axis of rotation*. Objects where the mass distribution is far from the axis of rotation have large values of $I$, while those where the mass is close to the axis of rotation have small values. 
Hopefully this makes sense, as it seems reasonable that it is harder to rotate an object when the same mass is placed further away from the axis of rotation, than when it is closer. 

In detail, $I$ is defined as 
$$ I = \sum_i m_i r_i^2,$$
where $m_i$ is the mass of each component of the system, and $r_i$ is its distance from the axis of rotation. 

For standard shapes such as a cylinder, we can perform this sum (or integral over a continuous mass distribution) and get an expression that we can use to calculate $I$. 
However in this particular experiment, we are dealing with a complicated shape, so rather than try to calculate $I$ from first principles, we will instead obtain the value from empirical measurement. 

In lecture we learn that if no external torques are present, angular momentum will be conserved. 
This is similar to the principle that if no external forces act in a collision, linear momentum will be conserved.
Note that for the rotating system there *can* be *forces* involved in the change of motion, but as long as those forces generate no *torque*, the angular momentum will be conserved.
Consider a system with initial rotational inertia $I_i$ and initial angular velocity $\omega_i$, which experiences a change  with no external torques, such that it has a final rotational inertia $I_f$ and final angular velocity $\omega_f$.
We will attempt to measure all four of these quantities, and from our measurements aim to test the hypothesis that angular momentum is conserved.

## Determining the Rotational Inertia

In principle for any system, the rotational inertia is obtained from $\sum_i m_i r_i^2$, where $m_i$ is the mass of each component of the system, and $r_i$ is its distance from the axis of rotation. 

For common shapes, this sum results in standard expressions that we can use. 
For example, for a cylinder of mass $M$ and radius $R$ being rotated around an axis through the cylinder's center and parallel to its length, the rotational inertia is $\frac{1}{2}MR^2$.
However, performing the sum can be challenging for more complicated shapes and objects. 
For this reason, we will use a method of direct measurement to obtain the rotational inertia for our apparatus.

### Calculating the Rotational Inertia from the Frequency of Small Oscillations

With the springs attached and the system in an equilibirum position, a small rotation of the chair away from equilibrium will cause it to oscillate. This is an example of Simple Harmonic Motion (SHM). You will study SHM later in the course. Below we will describe how to use our understanding of SHM to study the oscillation, and show how the rotational inertia is related to the frequency of small oscillations.
For now, if you have not encountered SHM before, then you can simply use the following relation:

   $$ I = \frac{2k R^2}{\omega_{osc}^2}$$


where here $\omega_{osc}$ is the *frequency of small oscillations* (in units of rad/sec) that you will measure.
(Note that $\omega_{osc}$ here is NOT the same as the angular velocity of a rotating object)
$k$ is the spring constant, as determined above. $R$ is the radius of the chair's trunk where the string/spring system attaches to - you will need to measure the diameter of this using calipers.

#### Derivation of Relationship between Rotational Inertia and Frequency of Small Oscillations

For reference, 
we will now proceed to derive how the rotational inertia of the system can be related to the measured oscillation frequency.
As stated above, you may skip this derivation for now and just use the formula quoted above in your calculation of rotational inertia. 
But I suggest you do return and review this derivation later, as it is a useful example of SHM in action. 

Assume the person+weights+chair+springs system is initially in the equilibrium position. Now consider displacing the chair by a small angle $\theta$ away from equilibrium. Therefore a point on the circumference where the spring is attached will move by a linear distance $R\theta$, where $R$ is the radius of the spring attachment point relative to the chair's axis rotation (this is about 4 inches for the rotating chair that we will use). 

For a small angular displacement, the change in length of the spring will therefore also be $\Delta x = R\theta$, and hence the spring will exert a Hooke's Law force of $k \Delta x = k R \theta$.
This force is applied at a distance $R$ from the axis of rotation, and is perpendicular to the radial direction, so the *torque* exerted by this spring is $k \Delta x R = k R^2 \theta$
There are two springs, at diametrically opposite points across the diameter, and the compression/extension forces result in *both torques being applied in the same direction*, so the magnitude of the total torque $T_{net}$ on the object is $2k R^2 \theta$. Whatever the direction of the initial displacement, the torque will be in the opposite sense, so that it tries to restore the system to equilibrium - thus, we take the torque to be negative. 

This total torque $T_{net}$ will result in an angular acceleration $\alpha$ of the system, where $T_{net} = I \alpha$, and $I$ is the *rotational inertia* of the system. 
Thus in this system, $-2k R^2 \theta =  I \alpha$
 and it follows that $$ \alpha = -(2k R^2/I) \theta$$
 
  This is an equation of Simple Harmonic Motion (SHM). The solution to this equation is an oscillation at some frequency  such that the angle at any time $\theta(t)$ is $\theta(t)=\theta_0 \cos\omega_{osc} t$, where $\omega_{osc}$ is the frequency of the oscillation, in units of radians per second.
  The angular acceleration $\alpha$ is the second derivative of $\theta(t)$ with respect to time. Thus $\alpha = -\omega_{osc}^2 \theta$ and by comparison with the expression above, we can conclude that $2k R^2/I = \omega_{osc}^2$ 
  and hence that the rotational inertia $I$ and the frequency of small oscillations $\omega_{osc}$ (in rad/s) are related by
  $$ I = \frac{2k R^2}{\omega_{osc}^2}$$
  
  Once we obtain the values for $k$ and $R$, then we can get $I$ by measuring $\omega_{osc}$ of the oscillations.
